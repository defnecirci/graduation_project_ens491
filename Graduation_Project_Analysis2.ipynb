```python
from sklearn.decomposition import PCA
def feature_importance(train_features,n_comp=2, k =6):
  model = PCA(n_components=n_comp).fit(train_features) # V, E 5
  # V -> feature sayisi tane column ve row
  # first column in V -> [1,2,5,7,8,1,2,5]
  # second column in V -> [. . .. . ]
  n_pcs = model.components_.shape[0]
  weighted_features = np.zeros((len(train_features.columns)))

  for i in range(len(weighted_features)):
    for j in range(len(weighted_features)):
      weighted_features[j] += np.abs(model.components_[i][j]) * model.explained_variance_[i]

  idx = np.argpartition(weighted_features, k)[:k]

  initial_feature_names = train_features.columns

  least_important_features = initial_feature_names[idx]
  return least_important_features
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from scipy.sparse.linalg import eigs

#drop_some_features = ['Day 7 Compressive Strength', 'Day 90 Compressive Strength', 'Day 50 Compressive Strength', 'Day 2 Compressive Strength']
#for d in drop_some_features:
    #if d in cement_data_numVarOnly:
    #  cement_data_numVarOnly.drop([d], axis=1, inplace=True)
# Get labels
if 'Day 28 Compressive Strength' in cement_data_numVarOnly.columns:
  data_labels = cement_data_numVarOnly.pop('Day 28 Compressive Strength')

cement_data_numVarOnly=cement_data_numVarOnly.fillna(cement_data_numVarOnly.mean())
print(cement_data_numVarOnly.shape)
```

## **3. Machine Learning**

**Linear Regression Model**



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

import matplotlib.pyplot as plt

r_states = range(100)
k_s = range(len(cement_data_numVarOnly.columns))
k_errors_test = np.zeros((len(cement_data_numVarOnly.columns)))
k_errors_train = np.zeros((len(cement_data_numVarOnly.columns)))
for k_num in k_s:
    normalized_cement_data_numVarOnly=(cement_data_numVarOnly-cement_data_numVarOnly.min())/(cement_data_numVarOnly.max()-cement_data_numVarOnly.min())
    least_important_features = feature_importance(normalized_cement_data_numVarOnly, len(normalized_cement_data_numVarOnly.columns), k=k_num)
    normalized_cement_data_numVarOnly.drop(least_important_features, axis=1, inplace=True)
    for state in r_states:
        x_train, x_test, y_train, y_test = train_test_split(normalized_cement_data_numVarOnly, data_labels, test_size=5, random_state=state)
    
        lin_reg = LinearRegression()
        lin_reg.fit(x_train, y_train)
        y_pred_test = lin_reg.predict(x_test)
        y_pred_train = lin_reg.predict(x_train)
        k_errors_test[k_num] += mean_squared_error(y_test, y_pred_test)
        k_errors_train[k_num] += mean_squared_error(y_train, y_pred_train)
        

k_errors_test /= len(r_states)
k_errors_train /= len(r_states)
best_k = np.argmin(k_errors_test)

print(np.min(k_errors_test))
print(np.min(k_errors_train))
print("best k", best_k)
```

```python
normalized_cement_data_numVarOnly=(cement_data_numVarOnly-cement_data_numVarOnly.min())/(cement_data_numVarOnly.max()-cement_data_numVarOnly.min()) # normalize
least_important_features = feature_importance(normalized_cement_data_numVarOnly, len(normalized_cement_data_numVarOnly.columns), k=best_k)
normalized_cement_data_numVarOnly.drop(least_important_features, axis=1, inplace=True)
x_train, x_test, y_train, y_test = train_test_split(normalized_cement_data_numVarOnly, data_labels, test_size=5, random_state=np.random.randint(0, 500))
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred_test = lin_reg.predict(x_test)
y_pred_train = lin_reg.predict(x_train)
print('Coefficients: \n', lin_reg.coef_)
print('Bias:', lin_reg.intercept_)
print("test mse:", mean_squared_error(y_test, y_pred_test))
print("train mse:", mean_squared_error(y_train, y_pred_train))
```
